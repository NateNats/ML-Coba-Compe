```python
features_col = data.columns.tolist()[:-1]
target_col = data.columns.tolist()[-1]
features_col
```

```python
target_col
```

```python
n_row, n_col = 10, 4
fig, axs = plt.subplots(n_row, n_col, figsize=(20, 60))
i, j = 0, 0

for n, col in enumerate(features_col):
    sns.boxplot(data=data, x=target_col, y=features_col[n], ax=axs[i, j])
    if j == n_col-1:
        j = 0
        i += 1
    else:
        j += 1
```

```python
df = data.copy()
df['DC201'][df['DC201'] == 'Layak Minum'] = 0
df['DC201'][df['DC201'] == 'Tidak Layak Minum'] = 1

corr = df.dropna().corr()
plt.figure(figsize=(25, 25))
sns.heatmap(corr, annot=True)
```

```python
target_list = df[target_col].value_counts()
target_list = target_list.index.tolist()
target_list
```

```python
n_row, n_col = 10, 4
fig, axs = plt.subplots(n_row, n_col, figsize=(20, 20))
i, j = 0, 0

for col in (features_col):
    plt.title(f'Distribution of {col} value')
    sns.histplot(x=df[col], hue=df['DC201'], ax=axs[i, j])
    if j == n_col-1:
        j = 0
        i += 1
    else:
        j += 1
plt.tight_layout()
```
